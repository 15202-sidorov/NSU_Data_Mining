```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
from sklearn import svm
```

```python

xx, yy = np.meshgrid(np.linspace(-3, 3, 500),
                     np.linspace(-3, 3, 500))
np.random.seed(0)
X = np.random.randn(300, 2)
Y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)

# fit the model
clf = svm.SVC()
clf.fit(X, Y)

# plot the decision function for each datapoint on the grid
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(figsize=(20, 5))
plt.subplot(1,4,3)
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',vmin=-abs(Z).max(),vmax=abs(Z).max(),
           origin='lower', cmap=plt.cm.PuOr_r)
contours = plt.contour(xx, yy, Z, levels=[0], linewidths=2,
                       linetypes='--')
plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired)

# clf = svm.SVC(kernel='linear')
# clf.fit(X, Y)
# Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
# Z = Z.reshape(xx.shape)

# plt.subplot(1,4,1)
# plt.imshow(Z, interpolation='nearest',
#            extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',vmin=-abs(Z).max(),vmax=abs(Z).max(),
#            origin='lower', cmap=plt.cm.PuOr_r)
# contours = plt.contour(xx, yy, Z, levels=[0], linewidths=2,
#                        linetypes='--')
# plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired)
# clf = svm.SVC(kernel='poly',degree=2)
# clf.fit(X, Y)
# Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
# Z = Z.reshape(xx.shape)

# plt.subplot(1,4,2)

# plt.imshow(Z, interpolation='nearest',
#            extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',vmin=-abs(Z).max(),vmax=abs(Z).max(), #norm=plt.Normalize,
#            origin='lower', cmap=plt.cm.PuOr_r)
# contours = plt.contour(xx, yy, Z, levels=[0], linewidths=2,
#                        linetypes='--')
# plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired)
# clf = svm.SVC(kernel='sigmoid')
# clf.fit(X, Y)
# Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
# Z = Z.reshape(xx.shape)
# plt.subplot(1,4,4)
# plt.imshow(Z, interpolation='nearest',
#            extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',vmin=-abs(Z).max(),vmax=abs(Z).max(),
#            origin='lower', cmap=plt.cm.PuOr_r)
# contours = plt.contour(xx, yy, Z, levels=[0], linewidths=2,
#                        linetypes='--')
# plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired)
```

```python

n = 100
a = np.random.normal(loc=-2, scale=1, size=(n*4,1))
b = np.random.normal(loc=2, scale=1, size=(n*4,1))
c = 5*np.random.randn(100,2)
#print np.hstack([a[:n],a[n:2*n]]).shape
X = np.vstack([np.hstack([a[:n],a[n:2*n]]),np.hstack([b[:n],b[n:2*n]]),c,np.hstack([a[2*n:3*n],b[2*n:3*n]]),np.hstack([b[3*n:],a[3*n:]])])
print X.shape
Y = np.hstack([np.zeros(2*n+50), np.ones(2*n+50)])
print Y.shape
ma=np.amax(X,axis=0)
mi=np.amin(X,axis=0)


xx, yy = np.meshgrid(np.linspace(2*mi[0], 2*ma[0], 500),
                     np.linspace(2*mi[1], 2*ma[1], 500))
# fit the model
clf = svm.SVC(probability=True,gamma=0.01)
clf.fit(X, Y)

# plot the decision function for each datapoint on the grid
Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])
Z = Z[:,1].reshape(xx.shape)
plt.figure(figsize=(20, 5))
plt.subplot(1,4,3)
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',vmin=0,vmax=1,
           origin='lower', cmap=plt.cm.PuOr_r)
contours = plt.contour(xx, yy, Z, levels=[0], linewidths=2,
                       linetypes='--')
plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired)

clf = svm.SVC(kernel='linear', probability=True)
clf.fit(X, Y)
Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])
Z = Z[:,1].reshape(xx.shape)

plt.subplot(1,4,1)
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',vmin=0,vmax=1,
           origin='lower', cmap=plt.cm.PuOr_r)
contours = plt.contour(xx, yy, Z, levels=[0], linewidths=2,
                       linetypes='--')
plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired)
clf = svm.SVC(kernel='poly',degree=2, probability=True)
clf.fit(X, Y)
Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])
Z = Z[:,1].reshape(xx.shape)

plt.subplot(1,4,2)

plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',vmin=0,vmax=1, 
           origin='lower', cmap=plt.cm.PuOr_r)
contours = plt.contour(xx, yy, Z, levels=[0], linewidths=2,
                       linetypes='--')
plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired)
clf = svm.SVC(kernel='sigmoid',gamma=0.1,coef0=1, probability=True)
clf.fit(X, Y)
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
#print Z.shape
Z1 = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])
#print Z1
Z_ = Z1[:,1].reshape(xx.shape)
plt.subplot(1,4,4)
plt.imshow(Z_, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',vmin=0,vmax=1,
           origin='lower', cmap=plt.cm.PuOr_r)
contours = plt.contour(xx, yy, Z_, levels=[0], linewidths=2,
                       linetypes='--')
plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired)
```
